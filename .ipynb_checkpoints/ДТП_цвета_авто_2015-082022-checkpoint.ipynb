```python
import pandas as pd
import numpy as np
import json
from flatten_json import flatten
import matplotlib.pyplot as plt
import seaborn as sns
```

# Готовим данные

```python
df = pd.read_json('../traffic_accident_colors/moskva.geojson')
```

```python
df = pd.json_normalize(
    df['features']
)
```

```python
df.head(2)
```

```python
#функция для проверки основных параметров датафрейма
def check(df):
    l=[]
    columns=df.columns
    for col in columns:
        dtypes=df[col].dtypes
        nunique=df[col].nunique()
        sum_null=df[col].isnull().sum()
        l.append([col,dtypes,nunique,sum_null])
    df_check=pd.DataFrame(l)
    df_check.columns=['column','dtypes','nunique','sum_null']
    return df_check
```

```python
vh = pd.json_normalize(
    df['properties.vehicles']
)
```

```python
vh1 = pd.json_normalize(
    vh[0]
)
vh1 = vh1[['brand', 'model', 'year', 'color']]
```

```python
vh2 = pd.json_normalize(
    vh[1]
)
vh2 = vh2[['brand', 'model', 'year', 'color']]
```

```python
vh3 = pd.json_normalize(
    vh[2]
)
vh3 = vh3[['brand', 'model', 'year', 'color']]
```

```python
vh4 = pd.json_normalize(
    vh[3]
)
vh4 = vh4[['brand', 'model', 'year', 'color']]
```

```python
vh5 = pd.json_normalize(
    vh[4]
)
vh5 = vh5[['brand', 'model', 'year', 'color']]
```

```python
vh6 = pd.json_normalize(
    vh[5]
)
vh6 = vh6[['brand', 'model', 'year', 'color']]
```

```python
vh7 = pd.json_normalize(
    vh[6]
)
vh7 = vh7[['brand', 'model', 'year', 'color']]
```

```python
vh8 = pd.json_normalize(
    vh[7]
)
vh8 = vh8[['brand', 'model', 'year', 'color']]
```

```python
vhdf = df.merge(vh1, left_index=True, right_index=True)
vhdf = vhdf.merge(vh2, left_index=True, right_index=True, suffixes=('_vh1', '_vh2'))
vhdf = vhdf.merge(vh3, left_index=True, right_index=True, suffixes=('_vh2', '_vh3'))
vhdf = vhdf.merge(vh4, left_index=True, right_index=True, suffixes=('_vh3', '_vh4'))
vhdf = vhdf.merge(vh5, left_index=True, right_index=True, suffixes=('_vh4', '_vh5'))
vhdf = vhdf.merge(vh6, left_index=True, right_index=True, suffixes=('_vh5', '_vh6'))
vhdf = vhdf.merge(vh7, left_index=True, right_index=True, suffixes=('_vh6', '_vh7'))
vhdf = vhdf.merge(vh8, left_index=True, right_index=True, suffixes=('_vh7', '_vh8'))
```

```python
vhdf['properties.datetime'] = vhdf['properties.datetime'].astype('datetime64')
```

```python
vhdf = vhdf[['properties.datetime', 'brand_vh1', 'model_vh1', 'year_vh1', 'color_vh1', 'brand_vh2',
       'model_vh2', 'year_vh2', 'color_vh2', 'brand_vh3', 'model_vh3',
       'year_vh3', 'color_vh3', 'brand_vh4', 'model_vh4', 'year_vh4',
       'color_vh4', 'brand_vh5', 'model_vh5', 'year_vh5', 'color_vh5',
       'brand_vh6', 'model_vh6', 'year_vh6', 'color_vh6', 'brand_vh7',
       'model_vh7', 'year_vh7', 'color_vh7', 'brand_vh8', 'model_vh8',
       'year_vh8', 'color_vh8']]
```

```python
vhdf = vhdf.rename(columns = {'properties.datetime':'dtp_datetime'})
```

```python
check(vhdf)
```

```python
vhdf["dtp_datetime"] = vhdf["dtp_datetime"].dt.to_period("Y")
```

```python
vhdf = vhdf.sort_values('dtp_datetime', ascending=False)
```

# Визуализация

```python
#Строим график с самыми аварийными цветами автомобилей с 2015-2022 год
ax = sns.countplot(x=vhdf['color_vh1'],
                   order=vhdf['color_vh1'].value_counts(ascending=False).index);
abs_values = vhdf['color_vh1'].value_counts(ascending=False).values
rel_values = vhdf['color_vh1'].value_counts(ascending=False, normalize=True).values * 100
lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]
sns.set(rc={'figure.figsize':(15,12)})
plt.title('График с самыми аварийными цветами автомобилей с 2015-2022 год')
plt.xlabel('Цвет автомобиля')
plt.ylabel('Количество ДТП')
plt.xticks(rotation=90)
ax.bar_label(container=ax.containers[0], labels=lbls)
plt.show()
```

```python
#График по годам
plt.rcParams["figure.figsize"]=[20,10]
sns.countplot(x="dtp_datetime", hue="color_vh1", data=vhdf, palette="Set1")
plt.title('График аварий с цветом автомобилей по годам')
plt.xlabel('Год ДТП')
plt.ylabel('Количество ДТП')
plt.draw()
```

```python
vhdf_table = vhdf[['dtp_datetime', 'color_vh1', 'brand_vh1']]\
.groupby(by=['dtp_datetime','color_vh1']).count().sort_values(['dtp_datetime', 'brand_vh1'], ascending=False)
```
